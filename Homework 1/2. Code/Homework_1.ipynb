### Retrieve the "Adj Close" column values as the daily prices of each stock

```python
import numpy as np
import pandas as pd

stocks = ['IBM', 'MSFT', 'GOOG', 'AAPL', 'AMZN', 'META']

stocks_df = pd.DataFrame()

for stock in stocks:
    stocks_df[stock] = pd.read_csv(f"https://raw.githubusercontent.com/ktxdev/AIM-5003/main/Homework%201/1.%20Data/{stock}.csv")['Adj Close']

stocks_df.head()
```

### Plot the daily price curve for each stock. So you will have 6 plots.

```python
from matplotlib import pyplot as plt

fig, axs = plt.subplots(2, 3, figsize=(20, 10))

for i, stock in enumerate(stocks_df.columns):
    r_idx = i // 3
    c_idx = i % 3
    axs[r_idx, c_idx].plot(stocks_df[stock])
    axs[r_idx, c_idx].set_title(f"Plot for {stock} Daily Prices")
```
